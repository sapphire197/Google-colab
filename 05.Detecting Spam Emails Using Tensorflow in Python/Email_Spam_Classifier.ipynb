```python
import numpy as np # numerical operations
import pandas as pd # data processing (data analysis and manipulation library)
import matplotlib.pyplot as plt # data visualization
from sklearn.model_selection import train_test_split # Split arrays or matrices into random train and test subsets.
from sklearn.feature_extraction.text import TfidfVectorizer # converts a collection of text documents into a matrix of TF-IDF features for text analysis.
from sklearn.linear_model import LogisticRegression # models the probability of a binary outcome using a logistic function
from sklearn.metrics import accuracy_score # measures the ratio of correctly predicted instances to the total instances in a classification task
```

```python
df = pd.read_csv('mail_data.csv') # reads the CSV file named mail_data.csv into a Pandas DataFrame df
```

```python
print(df)
```

```python
data = df.where((pd.notnull(df)),'') # replaces any NaN (missing) values in the DataFrame df with empty strings '', while keeping all non-null values unchanged.
```

```python
data.head()
```

```python
data.head(10) #first 10 rows
```

```python
data.info() # concise summary of the DataFrame data
```

```python
data.shape # returns a tuple representing the dimensions of the DataFrame data (rows, columns)
```

```python
data.loc[data['Category'] == 'spam', 'Category'] = 0
data.loc[data['Category'] == 'ham', 'Category'] = 1
```

```python
X = data['Message']
Y = data['Category']
```

```python
print(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3) # train=80%, test=20% (test_size=0.2)
```

```python
print(X.shape) # Total number of rows
print(X_train.shape) # 80% rows for training
print(X_test.shape) # 20% rows for testing
```

```python
print(Y.shape) # Total number of rows
print(Y_train.shape) # 80% rows for training
print(Y_test.shape) # 20% rows for testing
```

```python
feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english', lowercase=True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train)
```

```python
print(X_train_features)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train_features, Y_train) # Training LogicRegression with training data
```

```python
prediction_on_training_data = model.predict(X_train_features)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data) # Evaluate accuracy on training data
```

```python
print('Acc on training data: ', accuracy_on_training_data)
```

```python
prediction_on_test_data = model.predict(X_test_features)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data) # Evaluate accuracy on testing data
```

```python
print('Acc on testing data: ', accuracy_on_test_data)
```

```python
input_your_mail = ["This is the 2nd time we have tried to contact u. U have won the $400 prize. 2 claim is easy, just call 0870011881."]

input_data_features = feature_extraction.transform(input_your_mail)

prediction = model.predict(input_data_features)

print(prediction)

if prediction[0] == 0:
    print("Spam mail")
else:
    print("Ham mail")
```
